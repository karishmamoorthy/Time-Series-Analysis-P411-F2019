```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib nbagg
```

```python
x_axis = np.linspace(0, 1, 1000)
```

```python
uniform_raw = np.random.random_sample(1000) - 0.5

colours = ['red', 'yellow', 'green', 'cyan', 'blue', 'magenta', 'salmon', 'brown', 'black']
n = len(colours)

fig1, ax = plt.subplots(n, 1, sharex=True, figsize=(15, 12))
print(colours[0], " = 1 data process")
ax[0].set_title("Demonstration of Central Limit Theorem \nby summing over Uniform Random Distributions")
ax[0].hist(uniform_raw, histtype='step', bins=100, color=colours[0])
ax[0].grid(True)
for i in range(2, n+1):
    for j in range(2**i - 2**(i-1)):
        uniform_raw += np.random.random_sample(1000) - 0.5
    ax[i-1].hist(uniform_raw/np.sqrt(2**i), histtype='step', bins=100, color=colours[i-1])
    ax[i-1].grid(True)
    key = str(colours[i-1]) + " = ("+ str(2**(i-1)) + " data processes)/sqrt(" + str(2**(i-1)) + ")"
    print(key)
```
