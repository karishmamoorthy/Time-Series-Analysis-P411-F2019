```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib nbagg
```

```python
T = 1
```

```python
data_points = 1000
t = 1.5*T*np.linspace(-1, 1, data_points)
```

```python
series100 = np.zeros(data_points)
series10 = np.zeros(data_points)
series5 = np.zeros(data_points)
series2 = np.zeros(data_points)
series1 = np.zeros(data_points)

for m in range(1, 100):
    if m <= 1:
        series1 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series2 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series5 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series10 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series100 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
    elif m <= 2:
        series2 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series5 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series10 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series100 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
    elif m <= 5:
        series5 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series10 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series100 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
    elif m <= 10:
        series10 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
        series100 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
    else:
        series100 += 2*(1 - (-1)**m)*np.sin(2*m*np.pi*t/T)/(m*np.pi)
```

```python
plt.close()
plt.figure(figsize=(11, 7))
plt.plot(t, series1, 'r-', label="m = 1")
plt.plot(t, series2, 'b--', label="m = 2")
plt.plot(t, series5, 'g-', label="m = 5")
plt.plot(t, series10, 'y-', label="m = 10")
plt.plot(t, series100, 'k-', label="m = 100")
plt.title("Evolution of a Fourier series of a square wave")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.grid(True)
plt.legend()
```

```python
plt.close()
```
